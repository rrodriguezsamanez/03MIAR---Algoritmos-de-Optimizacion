## Torres de Hanoi
La técnica de **divide y vencerás** se basa en resolver un problema al descomponerlo en subproblemas del mismo tipo pero de menor tamaño. Estas soluciones parciales se combinan para obtener la solución del problema original. A medida que los subproblemas se hacen más pequeños, eventualmente se llega a casos simples o triviales que pueden resolverse fácilmente. Este enfoque utiliza comúnmente la **recursividad**.

En el caso de las **Torres de Hanoi**, el juego tradicional consta de tres varillas verticales y un conjunto de discos perforados por el centro. La cantidad de discos define el nivel de dificultad. Inicialmente, todos los discos están apilados en una de las varillas, ordenados de mayor a menor tamaño (de abajo hacia arriba). Las otras dos varillas comienzan vacías. El objetivo es trasladar todos los discos desde la varilla inicial a otra vacía, cumpliendo tres reglas:

1. Solo se puede mover un disco a la vez, y para mover uno, los demás deben estar en otros postes.
2. No se puede colocar un disco más grande sobre uno más pequeño.
3. Solo se puede mover el disco que esté en la parte superior de una pila.

Aunque existen diversas formas de llegar a la solución final, todas se basan en estas reglas y aplican estrategias alineadas con la técnica de divide y vencerás.


````python
#Torres de Hanoi - Divide y vencerás
def Torres_Hanoi(N, desde, hasta):
  #N - Cantidad de fichas
  #desde - torre inicio
  #hasta - torre fin
  if N == 1 : #condición de fin
    print("Lleva la ficha " ,desde , " hasta ", hasta )
      
  else: #condición recursiva
    Torres_Hanoi(N-1, desde, 6-desde-hasta )
    print("Lleva la ficha " ,desde , " hasta ", hasta )
    Torres_Hanoi(N-1, 6-desde-hasta  , hasta )
````

### Sucesión de Fibonacci

La **sucesión de Fibonacci** es una secuencia numérica en la que cada número es la suma de los dos anteriores. Comienza generalmente con los valores:



### Algoritmo visto en clase

````python
#Sucesión_de_Fibonacci
#Calculo del termino n-simo de la suscesión de Fibonacci
def Fibonacci(N:int):
  if N < 2:
    return 1
  else:
    return Fibonacci(N-1)+Fibonacci(N-2)

Fibonacci(8)
````

### Mejora de Algoritmo Fibonacci

````python
from functools import lru_cache

@lru_cache(maxsize=None)
def fibonacci_con_cache(n):
    if n < 2: return n
    return fibonacci(n - 1) + fibonacci(n - 2)
````

````python
start = time.time()
resultado = fibonacci_con_cache(valor)
end = time.time()
metodo = "Fibonacci-Cache"
print(f"[{metodo}] f(x) =>  f({valor}) → {resultado} | Tiempo: {end - start} segundos")
````

### Conclusiones sobre la optimización del algoritmo de Fibonacci

#### Comparación de rendimiento

Al calcular el término número 40 de la sucesión de Fibonacci, se observa una diferencia drástica en el tiempo de ejecución entre las dos implementaciones:

- **Fibonacci recursivo sin optimización**
  - Tiempo: ~16.44 segundos
  - Resultado: `165580141`
  - Costoso en tiempo, debido a la gran cantidad de llamadas recursivas redundantes (`O(2^n)`).

- **Fibonacci con `@lru_cache`**
  - Tiempo: ~0.00032 segundos
  - Resultado: `102334155`
  - Altamente eficiente gracias a la **memoización**, con complejidad `O(n)`.

#### Consideraciones sobre `@lru_cache`

Si bien el decorador `@lru_cache` mejora significativamente el rendimiento, también introduce una **consideración importante sobre el uso de memoria**:

- Cada llamada única a la función queda **almacenada en memoria**, lo que puede llevar a un consumo elevado si se calculan valores muy grandes de `n` o si se invoca la función muchas veces con distintos parámetros.
- En casos extremos, el uso intensivo del cache podría provocar **errores de memoria (`MemoryError`)**.

#### Recomendaciones

- Para cálculos aislados de términos moderadamente grandes (como `n < 10⁴`), `@lru_cache` es una excelente opción.
- Para valores muy grandes o aplicaciones que requieren eficiencia en memoria, se recomienda usar una versión **iterativa** que solo almacena los últimos dos valores:

## 2. Devolución de cambio por técnica voraz

La **técnica voraz** o **greedy algorithm** es una estrategia de diseño de algoritmos que construye una solución paso a paso, eligiendo en cada paso la opción que parece ser la mejor en ese momento. Esta técnica es especialmente útil en problemas de optimización, donde se busca minimizar o maximizar un valor, como el tiempo, el costo o, en este caso, el número de monedas.


Uno de los ejemplos clásicos para aplicar esta técnica es el problema de la **devolución de cambio**: dado un monto a devolver y un conjunto de denominaciones de monedas disponibles, se desea determinar la menor cantidad de monedas necesarias para devolver dicho monto.

El enfoque voraz consiste en tomar la moneda de mayor valor que no exceda el monto restante, restarla del total, y repetir el proceso hasta llegar a cero. Esta solución es eficiente y funciona correctamente **cuando las denominaciones cumplen ciertas propiedades**, como ser múltiplos entre sí (por ejemplo, monedas de 1, 2, 5, 10...).

Sin embargo, **no siempre garantiza una solución óptima** si el sistema de monedas es arbitrario. Aun así, en muchos casos prácticos como el sistema decimal de monedas, este método ofrece una solución rápida y suficientemente buena.

---

### Ejemplo:
Si se desea devolver 93 soles usando monedas/billetes de 1, 2, 5, 10, 20, 50, 100, 500 el algoritmo voraz seleccionaría:

- 1 moneda de 50  
- 2 moneda de 20  
- 1 moneda de 2  
- 1 monedas de 1  

**Total: 5 monedas**


````python
#Cambio de monedas aplicando el ejemplo anterior en algoritmos
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION
  print("No se puede calcular una solución")    
````

````python
SM = [500,100,50,20,10,5,2,1]
cambio_monedas(93,SM)
````

#### 6. Extender el algoritmo a 2D

````python

def fuerza_bruta_puntos_2D_mas_cercanos(puntos):
    distancia_minima = float('inf')
    puntos_mas_cercanos = None

    for i in range(len(puntos)):
        for j in range(i + 1, len(puntos)):
            distancia = math.dist(puntos[i], puntos[j])
            if distancia < distancia_minima:
                distancia_minima = distancia
                puntos_mas_cercanos = (puntos[i], puntos[j])

    return distancia_minima, puntos_mas_cercanos

def buscar_puntos_2D_mas_cercanos(puntos):
    puntos_ordenados = sorted(puntos, key=lambda punto: punto[0])
    return dividir_y_buscar_puntos_2D(puntos_ordenados)

def dividir_y_buscar_puntos_2D(puntos):
    cantidad = len(puntos)

    if cantidad <= 3:
        return fuerza_bruta_puntos_2D_mas_cercanos(puntos)

    mitad = cantidad // 2
    izquierda = puntos[:mitad]
    derecha = puntos[mitad:]

    dist_izquierda, puntos_izquierda = dividir_y_buscar_puntos_2D(izquierda)
    dist_derecha, puntos_derecha = dividir_y_buscar_puntos_2D(derecha)

    distancia_minima = min(dist_izquierda, dist_derecha)

    puntos_centrales = buscar_en_franja_central(puntos, distancia_minima)

    if puntos_centrales is not None:
        return distancia_minima, puntos_centrales

    if dist_izquierda < dist_derecha:
        return dist_izquierda, puntos_izquierda
    else:
        return dist_derecha, puntos_derecha

def buscar_en_franja_central(puntos, distancia_minima):
    cantidad = len(puntos)
    centro = cantidad // 2
    x_centro = puntos[centro][0]

    candidatos = [p for p in puntos if x_centro - distancia_minima <= p[0] <= x_centro + distancia_minima]
    candidatos.sort(key=lambda punto: punto[1])  # ordenar por Y

    mejores_puntos = None

    for i in range(len(candidatos)):
        for j in range(i + 1, min(i + 7, len(candidatos))):
            distancia = math.dist(candidatos[i], candidatos[j])
            if distancia < distancia_minima:
                distancia_minima = distancia
                mejores_puntos = (candidatos[i], candidatos[j])

    return mejores_puntos
````
